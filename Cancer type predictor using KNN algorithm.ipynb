```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split

path = "C:\\Users\\USER\\Desktop\\SriVarshniKS_Sapthagiricollege\\1)Cancertypepredictor_KNN\\PPT\\data.csv"

data=pd.read_csv(path)
data

inputs=data.drop('diagnosis','columns')
outputs=data.diagnosis
print(inputs)
print(outputs)

x_train,x_test,y_train,y_test = train_test_split(inputs,outputs,train_size=0.8)#train 80 percent
print(x_train)
print(x_test)
print(y_train)
print(y_test)

from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=13) #initializing my model
#train
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
print(y_pred)
print(y_test)

from sklearn.metrics import confusion_matrix
diagnosis=confusion_matrix(y_test,y_pred)
print(diagnosis)
```
